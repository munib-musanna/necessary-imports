```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Sample DataFrame
data = {
    'Category': np.random.choice(['A', 'B', 'C', 'D'], 100),
    'Value': np.random.randint(1, 100, 100),
    'Score': np.random.randn(100) * 15 + 50,
    'Gender': np.random.choice(['Male', 'Female'], 100),
    'Age': np.random.randint(18, 60, 100)
}
df = pd.DataFrame(data)
df
```

```python
# Basic Visualizations
sns.set(style="whitegrid")
```

```python
# 1. Bar Plot
plt.figure(figsize=(8, 5))
sns.barplot(x='Category', y='Value', data=df)
plt.title('Bar Plot')
plt.show()
```

```python
# 2. Count Plot
plt.figure(figsize=(8, 5))
sns.countplot(x='Category', hue='Gender', data=df)
plt.title('Count Plot')
plt.show()
```

```python
# 3. Box Plot
plt.figure(figsize=(8, 5))
sns.boxplot(x='Category', y='Score', data=df)
plt.title('Box Plot')
plt.show()
```

```python
# 4. Violin Plot
plt.figure(figsize=(8, 5))
sns.violinplot(x='Category', y='Score', data=df)
plt.title('Violin Plot')
plt.show()
```

```python
# 5. Strip Plot
plt.figure(figsize=(8, 5))
sns.stripplot(x='Category', y='Value', data=df, jitter=True)
plt.title('Strip Plot')
plt.show()
```

```python
# 6. Swarm Plot
plt.figure(figsize=(8, 5))
sns.swarmplot(x='Category', y='Value', data=df)
plt.title('Swarm Plot')
plt.show()
```

```python
# 7. Histogram
plt.figure(figsize=(8, 5))
sns.histplot(df['Score'], bins=20, kde=True)
plt.title('Histogram')
plt.show()
```

```python
# 8. KDE Plot
plt.figure(figsize=(8, 5))
sns.kdeplot(df['Score'], shade=True)
plt.title('KDE Plot')
plt.show()
```

```python
# 9. Scatter Plot
plt.figure(figsize=(8, 5))
sns.scatterplot(x='Age', y='Score', hue='Gender', data=df)
plt.title('Scatter Plot')
plt.show()
```

```python
# 10. Pair Plot
sns.pairplot(df, hue='Gender')
plt.suptitle('Pair Plot', y=1.02)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
categories_1 = ['A', 'B', 'C', 'D']
values_1 = [5, 15, 25, 35]

# Plot bar chart
plt.bar(categories_1, values_1)

# Add data labels
for index, value in enumerate(values_1):
    plt.text(index, value + 1, str(value), ha='center')

# Title and show plot
plt.title("Bar Chart with Data Labels")
plt.show()
```
